# Joining the Datasets

We have positives, alot of unlabeled and some negatives, with overlap. We need to combine the two. We now have 3 labels which we choose to label

* 1 - positive
* 0 - unlabeled
* -1 - negative

The negative data is just to be used for the test set. We also have overlapping (but not perfectly) data frames. As a result we choose to join the two dataframes, then drop the duplicates from the universe.

```python
features_cids = list(tech_features_formatted["cid"])
for cid in test_data_1_formatted["cid"]:
    if not cid in features_cids:
        print 'Not a perfect overlap, first missing cid is ', cid
        break
```

```python
td1_f = test_data_1_formatted
uni_f = tech_features_formatted

td1_f.loc[td1_f.label==0,'label']=-1  # set negatives to -1 (0 is unlabeled now)
```

```python
td1_uni = td1_f.append(uni_f, ignore_index=True)
td1_uni.describe()
```

And now we need to drop the duplicates

```python
td1_uni_f = td1_uni.drop_duplicates('cid')
td1_uni_f.describe()
```

Random shuffle so we can easily take smaller samples confidently

```python
_  = td1_uni_f.reindex(np.random.permutation(td1_uni_f.index))
```

```python
# check we havent lost anything
td1_pos = len(td1_f[td1_f['label'] == 1])
td1_neg = len(td1_f[td1_f['label'] == -1])
td1_uni_pos = len(td1_uni[td1_uni['label'] == 1])
td1_uni_neg = len(td1_uni[td1_uni['label'] == -1])

print td1_pos == td1_uni_pos
print td1_neg == td1_uni_neg
```

# Large Dataset

two files, one with labels and one without and a new universe

```python
large_data = pd.read_csv('../data/large_dataset.csv')
large_data.drop(["Unnamed: 0"], axis=1, inplace=True)
large_data.sort("cid", inplace=True)
large_data.head(20)
```

```python
large_data.describe()
```

```python
num_rows = len(large_data.index)
num_unique_cids = len(large_data.cid.unique())
print 'rows {}, uniques {}, difference {}'.format(num_rows, num_unique_cids, num_rows-num_unique_cids)
```

```python
large_data_duped = large_data.groupby("cid").agg(['count'])
lgd = large_data_duped['cid/fx/import_month_count']
duped_ids = lgd[lgd['count'] > 1]
duped_ids.head()
```

```python
large_data_labels = pd.read_csv('../data/large_dataset_labels.csv')
large_data_labels.drop(["Unnamed: 0"], axis=1, inplace=True)
large_data_labels.head(10)
print large_data_labels.label.value_counts()
print len(large_data_labels.index)
```

We need to deal with these duplicates, here 1 is positive, 0 is negative and -1 is an ambiguous outcome

```python
large_data_labels_unique = large_data_labels.drop_duplicates(subset='cid', take_last=True)
print large_data_labels_unique.label.value_counts()
print len(large_data_labels_unique.index)
```

```python
large_data_unique = large_data.drop_duplicates(subset='cid', take_last=True)
print len(large_data.index)
print len(large_data_labels_unique.index)
```

And the features

```python
large_data_w_labels = pd.merge(large_data_labels_unique, large_data_unique, on='cid', how='inner')
print len(large_data_w_labels.index)
print large_data_w_labels.label.value_counts()
large_data_w_labels.head(10)
```

Now loose the 0's (ambiguous) labels

```python
large_data_w_labels_pn = large_data_w_labels[large_data_w_labels.label >= 0]
print len(large_data_w_labels_pn.index)
print large_data_w_labels_pn.label.value_counts()
large_data_w_labels_pn.head(10)
```

Now we want to relabel the negatives (0) to -1

```python
ld_df = large_data_w_labels_pn
print ld_df.label.value_counts()
ld_df.loc[ld_df.label==0,'label'] = -1  # set negatives to -1 (0 is unlabeled now)
print ld_df.label.value_counts()
```

```python
large_data_w_labels.to_csv("../data/large_test_formatted.csv", index=False)
```
