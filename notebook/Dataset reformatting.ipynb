# Joining the Datasets

We have positives, alot of unlabeled and some negatives, with overlap. We need to combine the two. We now have 3 labels which we choose to label

* 1 - positive
* 0 - unlabeled
* -1 - negative

The negative data is just to be used for the test set. We also have overlapping (but not perfectly) data frames. As a result we choose to join the two dataframes, then drop the duplicates from the universe.

```python
features_cids = list(tech_features_formatted["cid"])
for cid in test_data_1_formatted["cid"]:
    if not cid in features_cids:
        print 'Not a perfect overlap, first missing cid is ', cid
        break
```

```python
td1_f = test_data_1_formatted
uni_f = tech_features_formatted

td1_f.loc[td1_f.label==0,'label']=-1  # set negatives to -1 (0 is unlabeled now)
```

```python
td1_uni = td1_f.append(uni_f, ignore_index=True)
td1_uni.describe()
```

And now we need to drop the duplicates

```python
td1_uni_f = td1_uni.drop_duplicates('cid')
td1_uni_f.describe()
```

Random shuffle so we can easily take smaller samples confidently

```python
td1_uni_f.reindex(np.random.permutation(td1_uni_f.index))
```

```python
# check we havent lost anything
td1_pos = len(td1_f[td1_f['label'] == 1])
td1_neg = len(td1_f[td1_f['label'] == -1])
td1_uni_pos = len(td1_uni[td1_uni['label'] == 1])
td1_uni_neg = len(td1_uni[td1_uni['label'] == -1])

print td1_pos == td1_uni_pos
print td1_neg == td1_uni_neg
```
